```python
sns.countplot(x='Preferred Foot',data=df);
```

```python
player_salary = df[["Name", "Wage"]].sort_values(by="Wage", ascending=False).reset_index(drop=True)
player_salary[:10]
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    x = "Name",
    y = "Wage",
    data = player_salary[:15]
)
plt.xlabel("Player's Name")
plt.ylabel("Salary in Euro")
plt.xticks(rotation=90)
plt.title('Most Salary ',fontsize=16)
plt.show()
```

```python
plt.subplots(figsize=(6,6))
plt.pie(df["Preferred Foot"].value_counts(), labels=["Right","Left"],autopct="%1.1f%%")
plt.show()
```

```python
plt.subplots(figsize=(12,9))
sns.countplot(x='Position',data=df)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x=df['Potential'],order=df['Potential'].value_counts().index.sort_values(),palette='dark')
plt.xticks(rotation='vertical')
plt.xlabel('Potential',fontsize=12)
plt.ylabel('No: of players',fontsize=12)
plt.title('Number of players vs potential',fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x=df['Height'],order=df['Height'].value_counts().index.sort_values())
plt.xticks(rotation=90,fontsize=10)
plt.xlabel('Height',fontsize=12)
plt.ylabel('No: of Players',fontsize=12)
plt.title('Players height',fontsize=16)
plt.show()
```

```python
sns.countplot(df['Joined'])
plt.xticks(rotation=45,fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('Joined',fontsize=40)
plt.xlabel('count',fontsize=10)
plt.title('per year player joined:',fontsize=10,fontweight='bold')
plt.show()
```

```python
print(df['Age'].median())
sns.histplot(x='Age',data=df)
plt.show()
```
